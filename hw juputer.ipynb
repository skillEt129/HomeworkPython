```python
from sympy import *
```

```python
x = Symbol('x', real=True)
```

```python
y = x ** 2 - 6 * abs(x) + 8
y
```

Определить корни

```python
roots_y = solve(y)
roots_y
```

Построить График

```python
graphic = plot(y)
```

```python
dot = diff(y)
dot
```

```python
roots_dot = solve(dot)
roots_dot
```

```python
g_d = plot(d)
```

```python
roots_dot.append(roots_dot[0] * -1)
roots_dot.append(0)
roots_dot = sorted(roots_dot)
roots_dot
```

```python
def intervals(roots):
    intervals = []
    intervals.append([roots[0]-1, roots[0]])
    for i in range(len(roots)-1):
        r = [roots[i], roots[i+1]]
        intervals.append(r)
    intervals.append([roots[-1], roots[-1]+1])
    return intervals
```

Вычислить вершины

```python
print(f'Точки экстремума функции {roots_dot[0]}, {roots_dot[-1]}')
```

```python
for i in roots_dot:
    left_value = i-0.01 if i == 0 else i-0.01*abs(i)
    right_value = i+0.01 if i == 0 else i+0.01*abs(i)
    if d.evalf(subs={x: left_val}) < d.evalf(subs={x: i}) and d.evalf(subs={x: right_value}) > d.evalf(subs={x: i}):
        print(f'точка минимума ⋁ - "{i}"')
    else:
        print(f'точка максимума ⋀ - "{i}"')
```


Интервалы на которых функция возрастает/убывает

```python
for i in intervals:
    start_int = -oo if intervals.index(i) == 0 else i[0]
    end_int = oo if intervals.index(i) == len(intervals)-1 else i[1]
    if d.evalf(subs={x:(i[0] + i[1])/2}) > 0:
        print(f'⮝ Функция возрастает на промежутке {start_int, end_int}')
    else:
        print(f'⮟ Функция убывает на промежутке {start_int, end_int}')
        
```
